```python
# importação do pyplot
import matplotlib.pyplot as plt

# plotando o gráfico de Vida util x Indice de saúde
plt.scatter(mvp['age'], mvp['bp'])

# incluindo título do gráfico e rótulos dos eixos
plt.title('Blood pressure x Age')
plt.xlabel('Age')
plt.ylabel('Blood pressure (mm/Hg)');
```

```python
# scatter plot com Seaborn
sns.scatterplot(x='age', y='bp', data=mvp);
```

```python
# Histograma

# plotando o histograma
plt.hist(mvp['age'], bins=2, edgecolor='black', color='yellow')

# incluindo título do gráfico e legenda
plt.title('Age')
plt.ylabel('Frequência')
plt.xlabel('age');
```

```python
# configurando um grupo de subplots em um grid de 1 linha e 2 colunas
# e também o tamanho da figura
fig, ax = plt.subplots(2,2, figsize = (8, 8))

# Convert 'wc' and 'rc' columns to numeric, coercing errors to NaN
mvp['wc'] = pd.to_numeric(mvp['wc'], errors='coerce')
mvp['rc'] = pd.to_numeric(mvp['rc'], errors='coerce')

# subplot1: histograma do Pressão arterial
ax[0,0].hist(mvp['bp'], bins=3, edgecolor='black')
ax[0,0].set_title("Blood pressure (mm/Hg)")

# subplot2: histograma da Idade
ax[0,1].hist(mvp['age'], bins=4, edgecolor='black')
ax[0,1].set_title("Age") # Changed title to reflect 'age'

# subplot3: histograma de células brancas
ax[1,0].hist(mvp['wc'].dropna(), bins=4, edgecolor='black')
ax[1,0].set_title("wc in cells/cumm")

# subplot4: histograma de células vermelhas
ax[1,1].hist(mvp['rc'].dropna(), bins=3, edgecolor='black');
ax[1,1].set_title("rc in millions/cmm");
```

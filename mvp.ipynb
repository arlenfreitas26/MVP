#Descrição do problema

A doença ou insificiência renal crônica caracteriza-se quando os rins perdem suas funções de maneira lenta e gradual. Segundo o ministério da saúde, doenças crônicas dos rins podem ter multiplas causas e multiplos fatores de risco.
Uma coleta de dados de uma determinada população permite a tentativa de correlacionar ou associar quais são as variáveis que mais contribuem para o diagnóstico crônico das perdas das funções renais. Esse projeto tem essa proposta, utilizar as informações obtidas nesse dataset para analisar as caracteristicas e quais delas associadas são mais potencializadoras para a comorbidade em questão.

**Hipótese do problema**
- Quais são as features que mais contribuem para estado crônico de mal funcionamento dos rins;

- Quais variáveis juntamente associadas geram mais predisposição aos pacientes a contrair um estado de mal fucionamento das suas funções renais.

**Tratamento de dados**

Nesse dataset tem muitas coluna que necessitam de tratar os valores das colunas com informações categóricas. Conforme pode-se observar na célula, há colunas com três ou quatro informações categóricas, considerando que deveriam exibir apenas dois valores. Por exemplo a coluna ***Coronary Artery Disease*** deveria exibir apena yes ou no ao invés de **no**, **yes** e **\tno**.

```python
#Exibir os valores nas colunas categóricas
#As colunas era para ter apenas dois tipos de dados, há necessidade de tratamento
mvp_cols = [col for col in mvp.columns if mvp[col].dtype == "object"]
for col in mvp_cols:
    unique_values = mvp[col].unique()
    print(f"Unique values in column '{col}':")
    print(unique_values)
    print("\n"+"="*60+"\n")
```

```python
#Substituindo \tyes e \tno por yes e no respectivamente
mvp['Diabetes Mellitus'] = mvp['Diabetes Mellitus'].replace([" yes", "\tyes"], "yes")
mvp['Diabetes Mellitus'] = mvp['Diabetes Mellitus'].replace([" no", "\tno"], "no")
mvp['Diabetes Mellitus'].unique()
```

```python
#Substituindo \tno por no.
mvp['Coronary Artery Disease'].replace(to_replace = {'\tno':'no'}, inplace = True)
mvp['Coronary Artery Disease'].unique()
```

```python
#Substituindo ckd\t por ckd.
mvp['Chronic Kidney Disease'] = mvp['Chronic Kidney Disease'].replace(to_replace = 'ckd\t', value = 'ckd')
mvp['Chronic Kidney Disease'].unique()
```

```python
# salvando um novo dataset para tratamento de missings

# recuperando os nomes das colunas
col = list(mvp.columns)

# o novo dataset irá conter todas as colunas do dataset original
mvpSemMissings = mvp[col[:]]

# substituindo os zeros por NaN
mvpSemMissings.replace(0, np.nan, inplace=True)

# exibindo visualização matricial da nulidade do dataset
ms.matrix(mvpSemMissings)
```

```python
# First, convert columns that should be numeric but are 'object' type to numeric,
# coercing errors to NaN. This handles problematic strings like '\t?' or '\t43'.
mvpSemMissings['Packed Cell Volume'] = pd.to_numeric(mvpSemMissings['Packed Cell Volume'], errors='coerce')
mvpSemMissings['White Blood Cells (cells/cmm)'] = pd.to_numeric(mvpSemMissings['White Blood Cells (cells/cmm)'], errors='coerce')
mvpSemMissings['Red Blood Cells (millions/cmm)'] = pd.to_numeric(mvpSemMissings['Red Blood Cells (millions/cmm)'], errors='coerce')

# Now, fill the NaN values with the median of each respective column.
mvpSemMissings['Packed Cell Volume'].fillna(mvpSemMissings['Packed Cell Volume'].median(), inplace=True)
mvpSemMissings['White Blood Cells (cells/cmm)'].fillna(mvpSemMissings['White Blood Cells (cells/cmm)'].median(), inplace=True)
mvpSemMissings['Red Blood Cells (millions/cmm)'].fillna(mvpSemMissings['Red Blood Cells (millions/cmm)'].median(), inplace=True)

# exibindo visualização matricial da nulidade do dataset
ms.matrix(mvpSemMissings)
```

```python
#Checando se há algum valor nulo
mvpSemMissings.isnull().sum()
```

```python
# filling null values

def random_value_imputation(feature):
    random_sample = mvpSemMissings[feature].dropna().sample(mvpSemMissings[feature].isna().sum(), replace=True)
    random_sample.index = mvpSemMissings[mvpSemMissings[feature].isnull()].index
    mvpSemMissings.loc[mvpSemMissings[feature].isnull(), feature] = random_sample

def impute_mode(feature):
    mode = mvpSemMissings[feature].mode()[0]
    mvpSemMissings[feature] = mvpSemMissings[feature].fillna(mode)
```

```python
# filling null values using random sampling method :

for col in mvpSemMissings:
    random_value_imputation(col)
```

```python
# Calculate the correlation list

# Create a copy of the DataFrame for encoding to preserve the original mvpSemMissings
mvpEncoded = mvpSemMissings.copy()

# Map categorical values to numerical representations
# For 'Chronic Kidney Disease': 'ckd' -> 1, 'notckd' -> 0
mvpEncoded['Chronic Kidney Disease'] = mvpEncoded['Chronic Kidney Disease'].map({'ckd': 1, 'notckd': 0})

# Map other binary categorical columns
mvpEncoded['Red Blood Cells'] = mvpEncoded['Red Blood Cells'].map({'normal': 1, 'abnormal': 0})
mvpEncoded['Pus Cells'] = mvpEncoded['Pus Cells'].map({'normal': 1, 'abnormal': 0})
mvpEncoded['Pus Cell Clumps'] = mvpEncoded['Pus Cell Clumps'].map({'present': 1, 'notpresent': 0})
mvpEncoded['Bacteria'] = mvpEncoded['Bacteria'].map({'present': 1, 'notpresent': 0})
mvpEncoded['Hypertension'] = mvpEncoded['Hypertension'].map({'yes': 1, 'no': 0})
mvpEncoded['Diabetes Mellitus'] = mvpEncoded['Diabetes Mellitus'].map({'yes': 1, 'no': 0})
mvpEncoded['Coronary Artery Disease'] = mvpEncoded['Coronary Artery Disease'].map({'yes': 1, 'no': 0})
mvpEncoded['Appetite'] = mvpEncoded['Appetite'].map({'good': 1, 'poor': 0})
mvpEncoded['Pedal Edema'] = mvpEncoded['Pedal Edema'].map({'yes': 1, 'no': 0})
mvpEncoded['Anemia'] = mvpEncoded['Anemia'].map({'yes': 1, 'no': 0})


# Now, calculate the correlation using the encoded DataFrame
target_corr = mvpEncoded.corr()['Chronic Kidney Disease'].abs().sort_values(ascending=False)[1:]
print(target_corr)
```

```python
# Create a bar chart to visualize the correlations
plt.figure(figsize=(10, 6))
sns.barplot(x=target_corr.index, y=target_corr.values)
plt.xticks(rotation=45, ha='right')
plt.xlabel('Features')
plt.ylabel('Correlation with diagnosis')
plt.title('Correlation between diagnosis and Features')
plt.tight_layout()
plt.show()
```

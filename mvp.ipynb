# Descrição do problema

A doença ou insificiência renal crônica caracteriza-se quando os rins perdem
suas funções de maneira lenta e gradual. Segundo o ministério da saúde, doenças
crônicas dos rins podem ter multiplas causas e multiplos fatores de risco.
Uma coleta de dados de uma determinada população permite a tentativa de correlacionar ou associar quais são as variáveis que mais contribuem para o diagnóstico crônico das perdas das funções renais. No que tange a natureza desse dataset, o conjuto de dados apresenta no contexto de aprendizado supervisionado, com intuito posterior de desenvolver modelos priditivos para prever doenças crônicas dos rins. Entretanto, a fase atual desse trabalho limita-se apenas a análise exploratória de dados (*Exploratory data analysis - EDA*), no pré-processamento dos dados e tratamento dos missings. Portanto tal projeta permite estabelecer relações qualitativas ou quantitativas dos atributos do dataframe com os quadros crônicos renais.

# **Tratamento de dados**

Nesse dataset há muitas colunas que necessitam de tratar os valores das colunas. Conforme pode-se observar na célula abaixo, há colunas com três ou quatro informações categóricas, considerando que deveriam exibir apenas dois valores. Por exemplo a coluna ***Coronary Artery Disease*** deveria exibir apena yes ou no ao invés de **no**, **yes** e **\tno**. Há também possíveis erros de digitação em colunas numéricas e dados faltantes (*missings*) que necessitam de avaliação e tratamento desses atributos.

```python
#Exibir os valores nas colunas categóricas
#As colunas era para ter apenas dois tipos de dados, há necessidade de tratamento
mvp_cols = [col for col in mvp.columns if mvp[col].dtype == "object"]
for col in mvp_cols:
    unique_values = mvp[col].unique()
    print(f"Unique values in column '{col}':")
    print(unique_values)
    print("\n"+"="*60+"\n")
```

```python
#Substituindo \tyes e \tno por yes e no respectivamente
mvp['Diabetes Mellitus'] = mvp['Diabetes Mellitus'].replace([" yes", "\tyes"], "yes")
mvp['Diabetes Mellitus'] = mvp['Diabetes Mellitus'].replace([" no", "\tno"], "no")
mvp['Diabetes Mellitus'].unique()
```

```python
#Substituindo \tno por no.
mvp['Coronary Artery Disease'].replace(to_replace = {'\tno':'no'}, inplace = True)
mvp['Coronary Artery Disease'].unique()
```

```python
#Substituindo ckd\t por ckd.
mvp['Chronic Kidney Disease'] = mvp['Chronic Kidney Disease'].replace(to_replace = 'ckd\t', value = 'ckd')
mvp['Chronic Kidney Disease'].unique()
```

```python
# Substituindo 'ckd' por 'yes' e 'notckd' por 'no'.
mvp['Chronic Kidney Disease'] = mvp['Chronic Kidney Disease'].replace({'ckd': 'yes', 'notckd': 'no'})
mvp['Chronic Kidney Disease'].unique()
```

# **Pré-processamento de Dados**

Limpeza e tratamento dos Dados: Verificação e eliminação de valores faltantes e erros nos dados. Procedeu-se a limpesa dos dados para manter a integridade e representatividade dos dados garantindo a acurancia nas análises.

```python
#Checando a quantidade missins no dataset
missing_values=mvp.isna().sum().sort_values(ascending=False)
print(missing_values)
```

```python
# salvando um novo dataset para tratamento de missings

# recuperando os nomes das colunas
col = list(mvp.columns)

# o novo dataset irá conter todas as colunas do dataset original
mvpSemMissings = mvp[col[:]]

# substituindo os zeros por NaN
mvpSemMissings.replace(0, np.nan, inplace=True)

# exibindo visualização matricial da nulidade do dataset
ms.matrix(mvpSemMissings)
```

```python
# Conversão das colunas que deveria ser numérica mas tem valores categóricos em numéricos, strings do tipo '\t?' or '\t43'.

mvpSemMissings['Packed Cell Volume'] = pd.to_numeric(mvpSemMissings['Packed Cell Volume'], errors='coerce')
mvpSemMissings['White Blood Cells (cells/cmm)'] = pd.to_numeric(mvpSemMissings['White Blood Cells (cells/cmm)'], errors='coerce')
mvpSemMissings['Red Blood Cells (millions/cmm)'] = pd.to_numeric(mvpSemMissings['Red Blood Cells (millions/cmm)'], errors='coerce')

# Substituindo NaN pela mediana.
mvpSemMissings['Packed Cell Volume'].fillna(mvpSemMissings['Packed Cell Volume'].median(), inplace=True)
mvpSemMissings['White Blood Cells (cells/cmm)'].fillna(mvpSemMissings['White Blood Cells (cells/cmm)'].median(), inplace=True)
mvpSemMissings['Red Blood Cells (millions/cmm)'].fillna(mvpSemMissings['Red Blood Cells (millions/cmm)'].median(), inplace=True)

# exibindo visualização matricial da nulidade do dataset
ms.matrix(mvpSemMissings)
```

```python
#Checando se há algum valor nulo
mvpSemMissings.isnull().sum()
```

```python
# removendo as colunas 'Sugar', 'Red Blood Cells' e 'Albumin'
mvpSemMissings.drop(['Sugar', 'Red Blood Cells', 'Albumin'], axis=1, inplace= True)

# exibindo visualização matricial da nulidade do dataset
ms.matrix(mvpSemMissings)
```

```python
# filling null values

def random_value_imputation(feature):
    random_sample = mvpSemMissings[feature].dropna().sample(mvpSemMissings[feature].isna().sum(), replace=True)
    random_sample.index = mvpSemMissings[mvpSemMissings[feature].isnull()].index
    mvpSemMissings.loc[mvpSemMissings[feature].isnull(), feature] = random_sample

def impute_mode(feature):
    mode = mvpSemMissings[feature].mode()[0]
    mvpSemMissings[feature] = mvpSemMissings[feature].fillna(mode)
```

```python
# filling null values using random sampling method :

for col in mvpSemMissings:
    random_value_imputation(col)
```

**Correlação dos atributos com o diagnóstico**

Com objetivo de correlacionar todos os atributos do dataframe com o atributos de diagnótico de doenças nos rins, foi necessário transformador os atributos com informação categórica em numérica de forma binária; por exemplo 1 (**um**) para casos positivos e 0 (**zero**) para casos negativos. Vide célula abaixo. As informações desse dataset corrobora com as informações obtidas obtidas nos sites da Sociedade Brasileira de nefrologia, Ministério da saúde e Organização Mundial de Saúde no que tange aos fatores de risco e diagnóstico. Analisando as correlações dos atributos com o diagnóstico de comprometimento das funções reansi de modo crônico, fatores como hipertenção e diabetes são os mais nocivos a saúde dos rins. Esses dois atributos, nos dados coletados desse dataframe, apresentam correlação em mais de 50% com diagnóstico de doença crônica dos rins.

```python
# Criando cópia do do data frame mvpSemMissings para preservá-lo

mvpEncoded = mvpSemMissings.copy()

if 'Age_Group' in mvpEncoded.columns:
    mvpEncoded = mvpEncoded.drop('Age_Group', axis=1)

# Transformando as informação do target em 1 quando yes e 0 quando no
mvpEncoded['Chronic Kidney Disease'] = mvpEncoded['Chronic Kidney Disease'].map({'yes': 1, 'no': 0})

# Transformando outros atributos em informação binária.
#mvpEncoded['Red Blood Cells'] = mvpEncoded['Red Blood Cells'].map({'normal': 1, 'abnormal': 0})
mvpEncoded['Pus Cells'] = mvpEncoded['Pus Cells'].map({'normal': 1, 'abnormal': 0})
mvpEncoded['Pus Cell Clumps'] = mvpEncoded['Pus Cell Clumps'].map({'present': 1, 'notpresent': 0})
mvpEncoded['Bacteria'] = mvpEncoded['Bacteria'].map({'present': 1, 'notpresent': 0})
mvpEncoded['Hypertension'] = mvpEncoded['Hypertension'].map({'yes': 1, 'no': 0})
mvpEncoded['Diabetes Mellitus'] = mvpEncoded['Diabetes Mellitus'].map({'yes': 1, 'no': 0})
mvpEncoded['Coronary Artery Disease'] = mvpEncoded['Coronary Artery Disease'].map({'yes': 1, 'no': 0})
mvpEncoded['Appetite'] = mvpEncoded['Appetite'].map({'good': 1, 'poor': 0})
mvpEncoded['Pedal Edema'] = mvpEncoded['Pedal Edema'].map({'yes': 1, 'no': 0})
mvpEncoded['Anemia'] = mvpEncoded['Anemia'].map({'yes': 1, 'no': 0})


# Calculando a correlação
target_corr = mvpEncoded.corr()['Chronic Kidney Disease'].abs().sort_values(ascending=False)[1:]
print(target_corr)
```

```python
# Exibindo a correlação dos atributos com a classificação de maneira gráfica

plt.figure(figsize=(10, 6))
sns.barplot(x=target_corr.index, y=target_corr.values)
plt.xticks(rotation=45, ha='right')
plt.xlabel('Atributos')
plt.ylabel('Correlação')
plt.title('Correlação dos atributos com diagnótico')
plt.tight_layout()
plt.show()
```
